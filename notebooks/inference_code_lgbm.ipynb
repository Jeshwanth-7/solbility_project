```python
import joblib 
```

```python


best_lgbm_pipeline = joblib.load("best_lgbm_pipeline.pkl")
   

X_test = joblib.load("X_test_random.pkl")

y_test = joblib.load("y_test_random.pkl")
```

```python
X_test.shape
```

```python
X_test.head()
```

```python
prediction_1 = best_lgbm_pipeline.predict(X_test)


print(prediction_1[:20])
```

```python
predictions = best_lgbm_pipeline.predict(X_test)
```

```python
import pandas as pd 

results =  pd.DataFrame({
    "Actual": y_test,
    "Predicted": predictions,
    "Error": abs(y_test - predictions)
})

results.head(20)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import numpy as np 

r2 = r2_score(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions) 
rmse = np.sqrt(mse)

print(f"R2 Score: {r2:.4f}")
print(f"Mean Absolute Error: {mae:.4f}")

print(f"Root Mean Squared Error: {rmse:.4f}")
```
