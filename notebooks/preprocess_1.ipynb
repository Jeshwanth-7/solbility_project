```python
from rdkit import Chem
from rdkit.Chem import SaltRemover
from rdkit.Chem.MolStandardize import rdMolStandardize

remover = SaltRemover.SaltRemover()

largest_fragment_chooser = (
    rdMolStandardize.LargestFragmentChooser()
)

def clean_molecule(smiles):

    try:

        mol = Chem.MolFromSmiles(smiles)

        if mol is None:
            return None

        mol = remover.StripMol(mol)

        mol = largest_fragment_chooser.choose(
            mol
        )

        Chem.SanitizeMol(mol)

        return mol

    except:
        return None

print("Before Cleaning:")
print(df.shape)

df["mol"] = df["Drug"].apply(
    clean_molecule
)

df = df[
    df["mol"].notnull()
].reset_index(drop=True)

print("\nAfter Cleaning:")
print(df.shape)
```

```python
df["clean_smiles"] = df["mol"].apply(
    Chem.MolToSmiles
)

df.head()
```

```python
import numpy as np

np.float = float
np.int = int
np.bool = bool
np.object = object

from mordred import Calculator, descriptors

calc = Calculator(
    descriptors,
    ignore_3D=True
)

descriptor_df = calc.pandas(
    df["mol"]
)

print("Descriptor Shape:")
print(descriptor_df.shape)
```

```python
descriptor_df = descriptor_df.apply(
    pd.to_numeric,
    errors="coerce"
)

print(descriptor_df.shape)

descriptor_df.head()
```

```python
missing_percent = (

    descriptor_df
    .isnull()
    .mean()
    * 100

)

missing_df = pd.DataFrame({

    "Missing_Percentage": missing_percent

}).sort_values(

    by="Missing_Percentage",
    ascending=False
)

print(missing_df.head(20))
```

```python
missing_threshold = 40

columns_to_keep = missing_percent[
    missing_percent <= missing_threshold
].index

descriptor_df = descriptor_df[
    columns_to_keep
]

print("Descriptor Shape After Missing Value Filtering:")
print(descriptor_df.shape)
```

```python
from rdkit.Chem import AllChem

def generate_fingerprint(mol):

    return list(

        AllChem.GetMorganFingerprintAsBitVect(
            mol,
            radius=2,
            nBits=1024
        )

    )

fingerprints = df["mol"].apply(
    generate_fingerprint
)

fingerprint_df = pd.DataFrame(
    fingerprints.tolist()
)

print("Fingerprint Shape:")
print(fingerprint_df.shape)
```

```python
fingerprint_df.columns = [

    f"FPMorgan_{i}"

    for i in range(
        fingerprint_df.shape[1]
    )
]
```

```python
import joblib

X_train = joblib.load(
    "X_train_random.pkl"
)

X_test = joblib.load(
    "X_test_random.pkl"
)

y_train = joblib.load(
    "y_train_random.pkl"
)

y_test = joblib.load(
    "y_test_random.pkl"
)
```

```python
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import StandardScaler

from sklearn.feature_selection import (
    VarianceThreshold,
    SelectKBest,
    f_regression
)
```

```python
preprocessor = Pipeline([

    (
        "imputer",
        SimpleImputer(strategy="median")
    ),

    (
        "variance_filter",
        VarianceThreshold(threshold=0.0)
    ),

    (
        "scaler",
        StandardScaler()
    ),

    (
        "feature_selection",
        SelectKBest(
            score_func=f_regression,
            k=1000
        )
    )

])
```

```python
from lightgbm import LGBMRegressor

from xgboost import XGBRegressor

from catboost import CatBoostRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.svm import SVR
```

```python
from sklearn.neural_network import MLPRegressor
```

```python

# MODEL PIPELINES


lgbm_pipeline = Pipeline([

    ("preprocessing", preprocessor),

    (
        "model",
        LGBMRegressor(
            n_estimators=300,
            random_state=42
        )
    )

])


xgb_pipeline = Pipeline([

    ("preprocessing", preprocessor),

    (
        "model",
        XGBRegressor(
            n_estimators=300,
            random_state=42
        )
    )

])


cat_pipeline = Pipeline([

    ("preprocessing", preprocessor),

    (
        "model",
        CatBoostRegressor(
            iterations=300,
            verbose=0,
            random_state=42,
            allow_writing_files=False
        )
    )

])


rf_pipeline = Pipeline([

    ("preprocessing", preprocessor),

    (
        "model",
        RandomForestRegressor(
            n_estimators=300,
            random_state=42,
            n_jobs=-1, 
            max_depth=6
        )
    )

])

svr_pipeline = Pipeline([

    ("preprocessing", preprocessor),

    (
        "model",
        SVR(
            kernel="rbf",
            C=10,
            epsilon=0.1
        )
    )

])

mlp_pipeline = Pipeline([
    
    ('preprocessor', preprocessor),
    (
        'model',MLPRegressor (
    
    
                hidden_layer_sizes = (256,128),
                activation = 'relu',
                solver = 'adam',
                alpha = 0.0001,
                batch_size = 'auto',
                learning_rate = 'adaptive',
                learning_rate_init = 0.0001,
                max_iter = 500,
                early_stopping= True,
                random_state = 42
        )
        
    )
    
])



# STORE ALL MODELS


models = {

    "LightGBM": lgbm_pipeline,

    "XGBoost": xgb_pipeline,

    "CatBoost": cat_pipeline,

    "RandomForest": rf_pipeline,
    
    "SVR": svr_pipeline,
    
    "MLpRegressor": mlp_pipeline

}
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import cross_val_score

results = []

for model_name, pipeline in models.items():

    cv_scores = cross_val_score(

        pipeline,

        X_train,

        y_train,

        cv=5,

        scoring="r2",

        n_jobs=-1

    )

    results.append({

        "Model": model_name,

        "CV Mean R2": cv_scores.mean(),

        "CV Std": cv_scores.std()

    })

   
    print(model_name)
   

    print("CV Scores:")
    print(cv_scores)

    print("\nMean CV R2:")
    print(cv_scores.mean())
```

```python
results_df = pd.DataFrame(results)

results_df = results_df.sort_values(

    by="CV Mean R2",

    ascending=False

)

results_df
```

```python
from sklearn.metrics import (

    r2_score,
    mean_absolute_error,
    mean_squared_error

)
```

```python
from sklearn.metrics import (

    r2_score,
    mean_absolute_error,
    mean_squared_error

)

test_results = []

for model_name, pipeline in models.items():

    # TRAIN MODEL
  

    pipeline.fit(
        X_train,
        y_train
    )



    
    # TRAIN PREDICTIONS
    

    y_train_pred = pipeline.predict(
        X_train
    )



    
    # TEST PREDICTIONS
    

    y_test_pred = pipeline.predict(
        X_test
    )



    
    # TRAIN METRICS
    

    train_r2 = r2_score(
        y_train,
        y_train_pred
    )

    train_rmse = np.sqrt(
        mean_squared_error(
            y_train,
            y_train_pred
        )
    )



    
    # TEST METRICS
    

    test_r2 = r2_score(
        y_test,
        y_test_pred
    )

    test_rmse = np.sqrt(
        mean_squared_error(
            y_test,
            y_test_pred
        )
    )

    test_mae = mean_absolute_error(
        y_test,
        y_test_pred
    )



    
    # STORE RESULTS
    

    test_results.append({

        "Model": model_name,

        "Train_R2": train_r2,

        "Test_R2": test_r2,

        "Train_RMSE": train_rmse,

        "Test_RMSE": test_rmse,

        "Test_MAE": test_mae

    })



    
    # PRINT RESULTS
    

    print("\n")
    print(model_name)
    

    print("Train R2 :", train_r2)
    print("Test R2  :", test_r2)

    print("Train RMSE :", train_rmse)
    print("Test RMSE  :", test_rmse)

    print("Test MAE :", test_mae)
```

```python
results_df = pd.DataFrame(
    test_results
)

results_df = results_df.sort_values(

    by="Test_R2",

    ascending=False

)

results_df
```
