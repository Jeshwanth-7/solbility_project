```python
from IPython.display import display_html
import pandas as pd

# ==========================================================
# HELPER
# ==========================================================

def create_table(df, pipeline_name):

    temp = df.copy()

    # Flatten MultiIndex columns
    temp.columns = [c[1] for c in temp.columns]

    temp["Model"] = temp.index
    temp["Pipeline"] = pipeline_name

    temp = temp.reset_index(drop=True)

    return temp


# ==========================================================
# RANDOM SPLIT
# ==========================================================

random_results = pd.concat([
    create_table(random_df, "Baseline"),
    create_table(robust_df, "Robust"),
    create_table(compact_df, "Compact")
], ignore_index=True)

random_results = random_results[
    [
        "Pipeline",
        "Model",
        "Test R2",
        "Test RMSE",
        "Test MAE",
        "CV Mean R2",
        "CV Std"
    ]
]

random_results = random_results.sort_values(
    by=["Test R2", "CV Mean R2"],
    ascending=[False, False]
).reset_index(drop=True)

random_results.insert(
    0,
    "Rank",
    range(1, len(random_results) + 1)
)

# ==========================================================
# STRATIFIED SPLIT
# ==========================================================

stratified_results = pd.concat([
    create_table(stratified_random_df, "Baseline"),
    create_table(stratified_robust_df, "Robust"),
    create_table(stratified_compact_df, "Compact")
], ignore_index=True)

stratified_results = stratified_results[
    [
        "Pipeline",
        "Model",
        "Test R2",
        "Test RMSE",
        "Test MAE",
        "CV Mean R2",
        "CV Std"
    ]
]

stratified_results = stratified_results.sort_values(
    by=["Test R2", "CV Mean R2"],
    ascending=[False, False]
).reset_index(drop=True)

stratified_results.insert(
    0,
    "Rank",
    range(1, len(stratified_results) + 1)
)

# ==========================================================
# STYLING
# ==========================================================

random_html = (
    random_results.style
    .background_gradient(
        cmap="RdYlGn",
        subset=["Test R2", "CV Mean R2"]
    )
    .background_gradient(
        cmap="RdYlGn_r",
        subset=["Test RMSE", "Test MAE", "CV Std"]
    )
    .format({
        "Test R2": "{:.4f}",
        "Test RMSE": "{:.4f}",
        "Test MAE": "{:.4f}",
        "CV Mean R2": "{:.4f}",
        "CV Std": "{:.4f}"
    })
    .hide(axis="index")
    .to_html()
)

stratified_html = (
    stratified_results.style
    .background_gradient(
        cmap="RdYlGn",
        subset=["Test R2", "CV Mean R2"]
    )
    .background_gradient(
        cmap="RdYlGn_r",
        subset=["Test RMSE", "Test MAE", "CV Std"]
    )
    .format({
        "Test R2": "{:.4f}",
        "Test RMSE": "{:.4f}",
        "Test MAE": "{:.4f}",
        "CV Mean R2": "{:.4f}",
        "CV Std": "{:.4f}"
    })
    .hide(axis="index")
    .to_html()
)

# ==========================================================
# DISPLAY SIDE BY SIDE
# ==========================================================

html = f"""
<div style="display:flex;gap:40px;align-items:flex-start">

<div style="width:50%">
<h2>🎲 Random Split Models</h2>
{random_html}
</div>

<div style="width:50%">
<h2>📊 Stratified Split Models</h2>
{stratified_html}
</div>

</div>
"""

display_html(html, raw=True)
```
